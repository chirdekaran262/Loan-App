# Deal with the missing values

```python
df.isnull().sum()
```

```python
df["LoanAmount"]=df["LoanAmount"].replace(np.NaN,df["LoanAmount"].mean())
```

```python
print(df["LoanAmount"])
```

```python
df.isnull().sum()
```

```python
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].replace(np.NaN,df["Loan_Amount_Term"].mean())
```

```python
df["Credit_History"]=df["Credit_History"].replace(np.NaN,df["Credit_History"].mean())
```

```python
df["Gender"].mode()
```

```python
df["Gender"]=df["Gender"].replace(np.NaN,"Male")
```

```python
df["Married"].mode()
```

```python
df["Married"]=df["Married"].replace(np.NaN,"Yes")
```

```python
df["Dependents"].mode()
```

```python
df["Dependents"]=df["Dependents"].replace(np.NaN,"0")
```

```python
df["Self_Employed"].mode()
```

```python
df["Self_Employed"]=df["Self_Employed"].replace(np.NaN,"No")
```
